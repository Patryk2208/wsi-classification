## Preprocessing of the "ortodoncja" dataset

### Steps
1. Handling missing data (no missing values in this dataset)
2. Handling outliers
3. Encoding categorical variables, so only the "growth direction" column
4. Splitting the dataset into train and test sets
5. Scaling the dataset

```python
from matplotlib import pyplot as plt
import numpy as np
from src.preprocessor import DataPreprocessor
import pandas as pd
import yaml

path_input  = '../data/raw/ortodoncja.csv'
path_output = '../data/processed/'
path_config = '../config/config.yaml'

df = pd.read_csv(path_input)
with open(path_config, 'r') as f:
    config = yaml.safe_load(f)

p = DataPreprocessor(df, config)
p.preprocess()


plt.figure(figsize=(12, 6))
plt.boxplot(p.X_train)
plt.xticks(range(1, p.X_train.shape[1]+1), rotation=45)
plt.title('Feature Distributions')
plt.tight_layout()
plt.show()
```
